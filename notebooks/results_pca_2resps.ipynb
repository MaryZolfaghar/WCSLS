```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
sns.set_style("darkgrid")
```

```python
ax_order = 'first'
ax_order_str = 'axF'
```

```python
with open('../../results/pca_results_mlp_2resps.P', 'rb') as f:
    mlp_results = pickle.load(f)   
# 
# with open('../../results/pca_%s_results_rnn.P' %(ax_order_str), 'rb') as f:
    # rnn_results = pickle.load(f)
```

```python
pca_mlp  = mlp_results['pca_results']
dist_mlp = mlp_results['dist_results']
# pca_rnn  = rnn_results['pca_results']
# dist_mlp = rnn_results['dist_results']
```

```python
# n_states = embed_2d_mlp.shape[0]
locs = [(i,j) for i in range(4) for j in range(4)]
dim1, dim2 = 0, 1
```

```python
# fig, axs = plt.subplots(1, 2, figsize=[10,4])
fig, axs = plt.subplots(1, 1, figsize=[7,6])
# plot embeddins
# mlp
embed_2d = pca_mlp['embed_2d']
ax = axs
ax.scatter(embed_2d[:, dim1], embed_2d[:, dim2], color='b')
for i, loc in enumerate(locs):
    text = str(loc)
    ax.annotate(text, (embed_2d[i, dim1], embed_2d[i, dim2]))
ax.set_title("MLP Face Embeddings", fontweight='bold')
ax.set_xlabel('dim%s' %(dim1))
ax.set_ylabel('dim%s' %(dim2))

# # rnn
# embed_2d = rnn_mlp['embed_2d']
# ax = axs[1]
# ax.scatter(embed_2d[:, dim1], embed_2d[:, dim2], color='b')
# for i, loc in enumerate(locs):
#     text = str(loc)
#     ax.annotate(text, (embed_2d[i, dim1], embed_2d[i, dim2]))
# ax.set_title("RNN Face Embeddings", fontweight='bold')
# ax.set_xlabel('dim%s' %(dim1))
# ax.set_ylabel('dim%s' %(dim2))


fig_str = '%s_pca_results_mlp_embeds' %(ax_order_str)
fig.suptitle('PCA Results', fontweight='bold')
plt.tight_layout()
fig.savefig(('../../figures/' + fig_str + '.jpg'), 
            bbox_inches = 'tight', pad_inches = 0)
```

```python
fig, axs = plt.subplots(1, 1, figsize=[7,6])
# plot embeddins
# mlp
avg_hidd_2d = pca_mlp['avg_hidd_2d']
ax = axs
ax.scatter(avg_hidd_2d[:, dim1], avg_hidd_2d[:, dim2], color='b')
for i, loc in enumerate(locs):
    text = str(loc)
    ax.annotate(text, (avg_hidd_2d[i, dim1], avg_hidd_2d[i, dim2]))
ax.set_title("MLP (avg) Hidden Embeddings", fontweight='bold')
ax.set_xlabel('dim%s' %(dim1))
ax.set_ylabel('dim%s' %(dim2))

# rnn
# embed_2d = pca_rnn['avg_hidd_2d']
# ax = axs[1]
# ax.scatter(avg_hidd_2d[:, dim1], avg_hidd_2d[:, dim2], color='b')
# for i, loc in enumerate(locs):
#     text = str(loc)
#     ax.annotate(text, (avg_hidd_2d[i, dim1], avg_hidd_2d[i, dim2]))
# ax.set_title("RNN (avg) Hidden Embeddings", fontweight='bold')
# ax.set_xlabel('dim%s' %(dim1))
# ax.set_ylabel('dim%s' %(dim2))

fig_str = '%s_pca_results_mlp_hidds_2resps' %(ax_order_str)
fig.suptitle('PCA Results', fontweight='bold')
plt.tight_layout()
fig.savefig(('../../figures/' + fig_str + '.jpg'), 
            bbox_inches = 'tight', pad_inches = 0)
```

```python
# fig, axs = plt.subplots(2, 3, figsize=[20,8])
fig, axs = plt.subplots(1, 3, figsize=[20,7])
# plot embeddins
# ------
# mlp
# ------
# plot pca context 0
ctx0_hidd_2d = pca_mlp['ctx0_hidd_2d']
ax = axs[0]
ax.scatter(ctx0_hidd_2d[:, dim1], ctx0_hidd_2d[:, dim2], color='b')
for i, loc in enumerate(locs):
    text = str(loc)
    ax.annotate(text, (ctx0_hidd_2d[i, dim1], ctx0_hidd_2d[i, dim2]))
ax.set_title("MLP face hidd. - context 0", fontweight='bold')

# plot context 1
ctx1_hidd_2d = pca_mlp['ctx1_hidd_2d']
ax = axs[1]
ax.scatter(ctx1_hidd_2d[:, dim1], ctx1_hidd_2d[:, dim2], color='b')
for i, loc in enumerate(locs):
    text = str(loc)
    ax.annotate(text, (ctx1_hidd_2d[i, dim1], ctx1_hidd_2d[i, dim2]))
ax.set_title("MLP face hidd. - context 1", fontweight='bold')

# plot both contexts
ctx_hidd_2d = pca_mlp['ctx_hidd_2d']
ax = axs[2]
half = int(ctx_hidd_2d.shape[0]/2) # the first half is for the context0
ax.scatter(ctx_hidd_2d[0:half, dim1], ctx_hidd_2d[0:half, dim2], color='r')
ax.scatter(ctx_hidd_2d[half:, dim1], ctx_hidd_2d[half:, dim2], color='b')
for i, loc in enumerate(locs):
    text = str(loc)
    ax.annotate(text, (ctx_hidd_2d[i, dim1], ctx_hidd_2d[i, dim2]))
    ax.annotate(text, (ctx_hidd_2d[i+half, dim1], ctx_hidd_2d[i+half, dim2]))
ax.set_title("MLP face hidd. - both contexts", fontweight='bold')

ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('dim%s' %(dim1))
ax.set_ylabel('dim%s' %(dim2))

fig_str = '%s_pca_results_mlp_hidds_ctx_2resps' %(ax_order_str)
fig.suptitle('PCA Results', fontweight='bold')
plt.tight_layout()
fig.savefig(('../../figures/' + fig_str + '.jpg'), 
            bbox_inches = 'tight', pad_inches = 0)
```

```python
fig, axs = plt.subplots(1, 2, figsize=[22,10])
# plot embeddins
# mlp
ctx_dim = 3
# plot hiddens - both contexts - no average over faces
hidd_2d = pca_mlp['hidd_2d']
samples_ctx = pca_mlp['samples_res']['samples_ctx']
ax = axs[0]
half = int(hidd_2d.shape[0]/2) # the first half is for the context0

ax.scatter(hidd_2d[0:half, dim1], hidd_2d[0:half, dim2], color='r')
ax.scatter(hidd_2d[half:, dim1], hidd_2d[half:, dim2], color='b')
for i  in range(half):
    ax.annotate(str(samples_ctx[i][ctx_dim][0].numpy()[0]), (hidd_2d[i, dim1], hidd_2d[i, dim2]))
    ax.annotate(str(samples_ctx[i+half][ctx_dim][0].numpy()[0]), (hidd_2d[i+half, dim1], hidd_2d[i+half, dim2]))
ax.set_title("MLP face hidd. - both contexts - no average", fontweight='bold')

# # plot both contexts - color coded by cong vs incong
ctx_dim = 4
incong_cong_2d = pca_mlp['incong_cong_2d']
samples_inc_c = pca_mlp['samples_res']['samples_inc_c']
ax = axs[1]
half = int(incong_cong_2d.shape[0]/2) # the first half is for the minor pairs
ax.scatter(incong_cong_2d[0:half, dim1], incong_cong_2d[0:half, dim2], color='r')
ax.scatter(incong_cong_2d[half:, dim1], incong_cong_2d[half:, dim2], color='b')
for i  in range(half):
    ax.annotate(str(samples_inc_c[i][ctx_dim][0].numpy()[0]), (incong_cong_2d[i, dim1], incong_cong_2d[i, dim2]))
    ax.annotate(str(samples_inc_c[i+half][ctx_dim][0].numpy()[0]), (incong_cong_2d[i+half, dim1], incong_cong_2d[i+half, dim2]))
ax.set_title("MLP face hidd. - cong vs incong pairs", fontweight='bold')

for ax in axs.reshape(-1): 
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel('dim%s' %(dim1))
    ax.set_ylabel('dim%s' %(dim2))

fig_str = '%s_pca_results_mlp_hidds_inc_c_ctx_2resps' %(ax_order_str)
fig.suptitle('PCA Results', fontweight='bold')
plt.tight_layout()
fig.savefig(('../../figures/' + fig_str + '.jpg'), 
            bbox_inches = 'tight', pad_inches = 0)
```
